### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

# III. Visualization

### Постройте гистограмму распределения признака 'total_bill'

```python
plt.figure(figsize= (7,5))
tips['total_bill'].hist(bins=30)
plt.title("Гистограмма распределения признака 'total_bill'")
plt.xlabel('Общий счет (total_bill)');
```

### Постройте scatterplot, представляющий взаимосвязь между признаками 'total_bill' и 'tip'

```python
plt.figure(figsize= (15,7))
plt.scatter(
    x = tips['total_bill'], y= tips['tip'], alpha=0.4, c = tips['tip'],
    s = 20 * tips['tip'], cmap='autumn', edgecolors = 'b', linewidths = 2
)
plt.title("Bзаимосвязь между признаками 'total_bill' и 'tip'")
plt.xlabel('Общий счет (total_bill)')
plt.ylabel('Чаевые (tip)')
#plt.legend()
plt.colorbar();
```

```python
plt.figure(figsize= (12,8))
sns.scatterplot(
    data=tips, x="total_bill", y="tip", hue="size", size="size",
    sizes=(20, 200), hue_norm=(0, 7), legend="full"
)
plt.title("Bзаимосвязь между признаками 'total_bill' и 'tip'")
plt.xlabel('Общий счет (total_bill)')
plt.ylabel('Чаевые (tip)');
```

### Постройте pairplot

```python
tips
```

```python
tips.loc[:, ['total_bill','tip','sex','smoker','size']];
gender_pal=dict(Female="violet",Male="darkblue")
print("\n\n Bзаимосвязь между признаками 'total_bill', 'tip', 'size' и полом посетителя \n")
sns.pairplot(tips.loc[:, ['total_bill','tip','sex','smoker','size']], hue="sex",palette=gender_pal, height=3);
```

### Постройте график взаимосвязи между признаками 'total_bill' и 'day'

```python
sns.set(style="ticks", palette="pastel")
f, axes = plt.subplots(1, 2, sharey=False, figsize=(16, 6))
sns.set(style="ticks", palette="muted")
sns.boxplot(x="day",y="total_bill" ,data=tips, ax=axes[0])
axes[0].set_title("График взаимосвязи между признаками 'total_bill' и 'day'")
sns.boxplot(x="day",y="total_bill",hue="sex",data=tips, palette=gender_pal, ax=axes[1])

axes[1].set_title("График взаимосвязи между признаками 'total_bill' и 'sex'")
sns.despine(offset=10, trim=True);
```

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
tips['time'].unique()
```

```python
f, axes = plt.subplots(1, 2, sharey=True, figsize=(14, 7))
sns.histplot(tips.query('time == "Lunch"')['tip'], kde=True, color="blue", ax=axes[0], bins=30).\
    set_title("Гистограмма распределения признака 'tip' (Lunch time)")
sns.histplot(tips.query('time == "Dinner"')['tip'], kde=True, color="purple", ax=axes[1], bins=30).\
    set_title("Гистограмма распределения признака 'tip' (Dinner time)")
sns.despine(offset=10, trim=False);
```

```python
print(tips.query('time == "Lunch"')['tip'].median()) #2,25
print(tips.query('time == "Dinner"')['tip'].median()) #3,0
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'
